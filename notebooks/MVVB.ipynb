This notebook provides all experiments and data manipulations for the paper *Folktale Stability and Transmission of Motifs*. Before we will delve into the matter, we will first do some basic data analyses to get a feeling of the data set. Let's start with reading the annotated data set which is a csv file.

```python
import pandas as pd
df = pd.read_csv("../data/motifs-vvb.csv", index_col=0)
df.describe()
```

The csv file contains for each folktale a number of features, such as its genre, the location where it was told or collected, its storytype (classified according to the *Internation Folktale Type Catalog* by Uther (2004)) etc.

As the following bar chart clearly shows, the annotated folktales are rather unevenly distributed over over the different genres. Most tales are fairy tales and only jokes and legends are relatively frequent as well.

```python
df.genre.value_counts().plot(kind='bar')
```

The ATU catalog also distinguishes a number of folktale genres. These genres are more content-based, such as "Animal Tales" or "Tales of Magic". It would be interesting to see how the annotated folktales over these genres as well.

To do so, we must first add a column "ATU genre" to the dataframe. Then for each folktale we will look up its ATU number and write down its corresponding ATU genre. [Wikipedia](http://en.wikipedia.org/wiki/Aarne%E2%80%93Thompson_classification_system) gives an index of all folktale types and their corresponsing ATU numbers. Before we can efficiently use that information, we must first transform the ATU numbers in the dataframe (now represented as strings, such as "ATU 0333") to actual integers. We'll do so as follows:

```python
import re

df['atunumber'] = df['storytype'].apply(lambda x: int(re.search('[0-9]+', x).group()))

df[['storytype', 'atunumber']].head()
```

Now that we have the ATU numbers represented as integers, we can add the ATU genre column to the dataframe:

```python
def between(x, a, b):
    return x >= a and x <= b

df['atugenre'] = df['atunumber'].apply(
    lambda x: 'ANIMAL TALES' if between(x, 1, 299) else
              'TALES OF MAGIC' if between(x, 300, 749) else
              'RELIGIOUS TALES' if between(x, 750, 849) else
              'REALISTIC TALES' if between(x, 850, 999) else
              'TALES OF THE STUPID OGRE' if between(x, 1000, 1199) else
              'ANECDOTES AND JOKES' if between(x, 1200, 1999) else
              'FORMULA TALES')

df[['storytype', 'atugenre']].head()
```

And plot:

```python
df["atugenre"].value_counts().plot(kind='bar')
```

Interestingly, the ATU genres show a different genre distribution than the genres as classified in the Dutch Folktale Database (DFD). Whereas fairy tales are most common in the DFD, in the ATU classification "Anecdotes and Jokes" are most common. 

The folktales in the DFD are collected from a range of different source. Some are transcriptions of oral tellings others are copies from literary versions. Let's have a look at the distribution of the annotated motif set over the different sources:

```python
df['tradition'].value_counts().plot(kind='bar')
```

The vast majority of folktales is collected either from books or from oral tellings. Next come magazines and letters which represent about 150 folktales each. As we will see below, the source or tradition of the retelling plays a very imporant part in the stability and variation of the folktale. 

Now, let's investigate the motif column. For each folktales we extracted a string of motifs in their order of appearance in the story. Motifs are separated by a semi colon. Before can actually do something with this data, we must first transform it into a more convenient format. There are basically two ways in which we can represent the motifs in the dataframe. For each motif listed for a particular tale, we duplicate all other fields, or we represent the motifs as binary features for each folktale (leading to quite a large table). For now I will take the first approach:

```python
motifs = df['motifs'].apply(lambda x: list(set(x.split(';')))).apply(pd.Series, 1).stack()
motifs.index = motifs.index.droplevel(-1)
motifs.name = 'motif'
motifs = df.join(motifs)
motifs[['idnumber', 'storytype', 'motif']].head()
```

Now that we have a new dataframe with each motif represented separately, let's first have a look at the distribution of these motifs:

```python
motifs['motif'].value_counts().plot()
```

Next let's zoom in on the 15 most frequently used motifs in the annotated dataset:

```python
motifs['motif'].value_counts()[:15].plot(kind='bar')
```

The three most frequently used motifs are Z41 *The old woman and her pig*, N211.1 *Lost ring found in fish* and F913 *Victims rescued from swallower's belly*. Many motifs are bound to particular stories and story types. It would therefore also be interesting to see which motifs are often reused in different stories, i.e. which motifs are not only syntagmatically recurrent but also paradigmatically?

A rather primitive, yet effective way to measure this would be to count for each motif in how many different folktale types it occurs:

```python
typedist = motifs.groupby('motif').apply(lambda x: len(set(x['storytype'])))
typedist.sort(ascending=False)
typedist.head(n=10)
```

Only a very few motifs occur over the borders of a particular storytype. L161 *Lowly hero marries princess* occurs in 8 different storytypes and F601 *Extraordinary companions* occurs in 6 different story types. As the following plot shows, the vast majority of motifs belong to a particular folktale type exclusively. 

```python
ax = typedist.value_counts().plot()
ax.set_xlabel("Number of different folktale types")
ax.set_ylabel("Number of items")
```

So, most motifs in the dataset are bound to a particular folktale type. However, within these types, that is considering all instances of a particlar type, we can observe quite an amount of variation in terms of which motifs are actually used. To give an example, I will construct a graph of all variants of ATU 330. Nodes are represented by the stories and the edges between nodes are formed if and only if two stories share a particular motif.

```python
import networkx as nx
from collections import defaultdict
from itertools import combinations

G = nx.Graph()

edges = defaultdict(list)
for id, row in motifs[motifs.atunumber == 330].iterrows():
    edges[row.motif].append(row.idnumber)
for motif, stories in edges.iteritems():
    for s1, s2 in combinations(stories, 2):
        G.add_edge(s1, s2, label=motif)
nx.write_graphml(G, "documents.graphml")
```

The two graphs show some nice prototype effects of either some stories being the most central or some motifs. The layout of the first graph (document based) provides some ground for the hypothesis that there are subgroups of stories or subtypes within a story type. To get a better picture of the relation or similarity between each story in the ATU 330 group, I we now perform a simple cluster analysis where we compute the pairwise similarity between each instance on the basis of the motifs it has in common with other instances.

The Python library [distance](https://pypi.python.org/pypi/Distance/0.1.3) provides a nice interface to a range of similarity metrics, such as Levenshtein distance and Jaccard distance. For now we will use the normalized Jaccard metric to compute all pairwise distances.

```python
import numpy as np
import distance as dist
from phd.utils import encoder

def distances_for_type(storytype, fn=dist.jaccard):
    encode = encoder()
    instances = df[df.storytype==storytype]
    matrix = np.zeros((instances.shape[0], instances.shape[0]))
    motifs = instances[['idnumber', 'motifs']]['motifs'].apply(lambda x: x.split(';'))
    for (_, s1), (_, s2) in combinations(
        pd.concat([instances['idnumber'], motifs], 1).iterrows(), 2):
        s1_id, s2_id = encode(s1.idnumber), encode(s2.idnumber)
        matrix[s1_id, s2_id] = fn(s1.motifs, s2.motifs)
        matrix[s2_id, s1_id] = matrix[s1_id, s2_id]
    labels = sorted(instances['idnumber'], key=encode)
    dm = pd.DataFrame(matrix, columns=labels, index=labels)
    return dm
```

The different dendrograms hint at a pattern where each folktale type is divided into two or more subtypes. The question naturally is what is reponsible for this variation? Also, although we can detect many dissimilarities, there are also many stable motifs. Which are they? Are their some common characteristics of these motifs to remain stable? Do they, for example, occupy a particular position in the story? Or are they of a particular nature? Such questions will be discussed below. 

Before we'll go into that, I first want to more quantitatively express the amount of variation in the different folktale types. Can we give it a number? There are of course many different ways in which we can measure the variation (e.g. entropy, variation coefficient). And these metrics are also dependent on the distance metric used (e.g. Jaccard distance or Levenshtein distance).

I will compare two distance metrics, Jaccard and Levenshtein. For each folktale type we will compute the pairwise distances using both distance metrics. Then, we will experiment with a couple of metrics to quantitatively express the variation within these types.

```python
from functools import partial

def avg_distance(dm):
    return dm.mean().mean()

variation = []
for storytype, count in df['storytype'].value_counts().iteritems():
    if count == 1: continue
    dm_jaccard = distances_for_type(storytype, dist.jaccard)
    dm_levenshtein = distances_for_type(storytype, partial(dist.levenshtein, normalized=True))
    variation.append((df[df.storytype==storytype]['atugenre'].values[0], 
                      storytype, avg_distance(dm_jaccard), avg_distance(dm_levenshtein)))

variation = pd.DataFrame(variation, columns=['atugenre', 'storytype', 'jaccard', 'levenshtein'])
```

```python
fig, axs = sb.plt.subplots(1, 2, figsize=(14, 4))

variation['jaccard'].plot(title='jaccard', kind='density', ax=axs[0])
variation['levenshtein'].plot(title='levenshtein', kind='density', ax=axs[1])
```

These two density plots appear to indicate that overall there is not to much variation present within folktale types. We do see some small bumps in the graphs. It would therefore be interesting to take a look at the variation at a level of more granularity. The following plots show the variation on the level of atu genres:

```python
fig, axs = sb.plt.subplots(1, 2, figsize=(14, 4))

_ = variation.groupby('atugenre')['jaccard'].plot(
        kind='density', ax=axs[0], legend=True, title='jaccard')
_ = variation.groupby('atugenre')['levenshtein'].plot(
        kind='density', ax=axs[1], legend=True, title='levenshtein')
```

To control for the possible negative levelling effect of averages, it is interesting to compare these plot to another method to compute the variation. This time we compute the variation in terms of the average distance to a prototype where a prototype is defined as the centroid of all individual stories. As can be seen from the plot below, the results are almost identical to the ones above. The strongest conclusion we can draw is that stories and folktale types within the ATU genre "Tales of Magic" demonstrate the highest degree of variation.

```python
from sklearn.preprocessing import LabelBinarizer

binarizer = LabelBinarizer()

variation = []
for storytype, count in df['storytype'].value_counts().iteritems():
    if count == 1: continue
    stories = df[df.storytype == storytype]
    stories = stories[['idnumber', 'motifs']]['motifs'].apply(
        lambda x: list(set(x.split(';')))).values
    stories = binarizer.fit_transform(stories)
    prototype = stories.mean(axis=0)
    distances = np.abs(stories - prototype).sum(axis=1).mean()
    variation.append((df[df.storytype==storytype]['atugenre'].values[0], 
                      storytype, distances.mean()))

variation = pd.DataFrame(variation, columns=['atugenre', 'storytype', 'distance'])
```

```python
fig, axs = sb.plt.subplots(1, 1, figsize=(8, 4))
_ = variation.groupby('atugenre')['distance'].plot(kind='density', legend=True)
```
