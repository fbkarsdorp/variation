To get a better picture of the relation or similarity between each story in the ATU 330 group, I we now perform a simple cluster analysis where we compute the pairwise similarity between each instance on the basis of the motifs it has in common with other instances.

The Python library [distance](https://pypi.python.org/pypi/Distance/0.1.3) provides a nice interface to a range of similarity metrics, such as Levenshtein distance and Jaccard distance. For now we will use the normalized Jaccard metric to compute all pairwise distances.

```python
import numpy as np
import distance as dist
from phd.utils import encoder

def distances_for_type(storytype, fn=dist.jaccard):
    encode = encoder()
    instances = df[df.storytype==storytype]
    matrix = np.zeros((instances.shape[0], instances.shape[0]))
    motifs = instances[['idnumber', 'motifs']]['motifs'].apply(lambda x: x.split(';'))
    for (_, s1), (_, s2) in combinations(
        pd.concat([instances['idnumber'], motifs], 1).iterrows(), 2):
        s1_id, s2_id = encode(s1.idnumber), encode(s2.idnumber)
        matrix[s1_id, s2_id] = fn(s1.motifs, s2.motifs)
        matrix[s2_id, s1_id] = matrix[s1_id, s2_id]
    labels = sorted(instances['idnumber'], key=encode)
    dm = pd.DataFrame(matrix, columns=labels, index=labels)
    return dm
```

Before we'll go into that, I first want to more quantitatively express the amount of variation in the different folktale types. Can we give it a number? There are of course many different ways in which we can measure the variation (e.g. entropy, variation coefficient). And these metrics are also dependent on the distance metric used (e.g. Jaccard distance or Levenshtein distance).

I will compare two distance metrics, Jaccard and Levenshtein. For each folktale type we will compute the pairwise distances using both distance metrics. Then, we will experiment with a couple of metrics to quantitatively express the variation within these types.

```python
from functools import partial

def avg_distance(dm):
    return dm.mean().mean()

variation = []
for storytype, count in df['storytype'].value_counts().iteritems():
    if count == 1: continue
    dm_jaccard = distances_for_type(storytype, dist.jaccard)
    dm_levenshtein = distances_for_type(storytype, partial(dist.levenshtein, normalized=True))
    variation.append((df[df.storytype == storytype]['atugenre'].values[0], 
                      storytype, avg_distance(dm_jaccard), avg_distance(dm_levenshtein)))

variation = pd.DataFrame(variation, columns=['atugenre', 'storytype', 'jaccard', 'levenshtein'])
variation.groupby('atugenre').mean().rank(ascending=False)
```

A fast way of doing this is to compute a ranking of motif stability for each story type by simple frequency counts.
```python
rankings = {}
for storytype, count in df['storytype'].value_counts().iteritems():
    if count == 1: continue
    stories = df[df.storytype == storytype]
    counts = stories['motifs'].map(
        lambda x: x.split(';')).apply(pd.Series, 1).stack().value_counts()
    # normalize
    counts = counts / stories.shape[0]
    rankings[storytype] = counts.to_dict()
```

```python
from collections import defaultdict

stable = defaultdict(lambda: [0, 0])
for storytype, motifs in sorted(rankings.items()):
    motif = max(motifs, key=motifs.__getitem__)
    stable[motif[0]][0] += 1
    for motif in motifs:
        stable[motif[0]][1] += 1
motifs, counts = zip(*stable.items())
d = pd.DataFrame(list(counts), index=motifs, columns=['stable', 'total'])
d['stable'] = d['stable'] / d['stable'].sum()
d['total'] = d['total'] / d['total'].sum()
d.plot(kind='bar')
```

Although not conclusive, the dendrograms above seem to hint at an influence of geographical distance on the variation of motifs observed within certain folktale types. A number of questions arise from these initial plots.

These and other questions we be central in the next part of this notebook. First we will define a function that can be used to compute the distance between locations. We want to make it a general function that is able to compute the distance on multiple levels (cities, provinces and even countries) using any user defined distance metric (e.g. Jaccard, Levenshtein, Needle Wunsch etc). Here it is:

```python
from phd.utils import memo
from functools import partial

def distance_between(loc_a, loc_b, df, level='place', fn=dist.jaccard):
    encode = encoder()
    loc_a_df = df[df[level]==loc_a]
    loc_b_df = df[df[level]==loc_b]
    distances = []
    for storytype in set.union(set(loc_a_df.storytype.values),
                               set(loc_b_df.storytype.values)):
        loc_a_stories = loc_a_df[loc_a_df.storytype==storytype]
        loc_b_stories = loc_b_df[loc_b_df.storytype==storytype]
        if loc_a_stories.empty or loc_b_stories.empty:
            distances.append(1.0)
        else:
            for _, s1 in loc_a_stories.iterrows():
                for _, s2 in loc_b_stories.iterrows():
                    distances.append(fn(s1.motifs, s2.motifs))
        # wil je per storytype een gemiddelde of alleen over het geheel?
    return sum(distances) / len(distances) if distances else np.NAN
```

Next we will define two utility functions that allow us to conveniently use both the jaccard and the levenshtein distance:

```python
def jaccard_location(loc_a, loc_b, df, level, **kwargs):
    def fn(s1, s2):
        return dist.jaccard(set(s1.split(';')), set(s2.split(';')), **kwargs)
    return distance_between(loc_a, loc_b, df, level, fn)

def levenshtein_location(loc_a, loc_b, df, level, **kwargs):
    def fn(s1, s2):
        return dist.nlevenshtein(s1.split(';'), s2.split(';'), method=2)
    return distance_between(loc_a, loc_b, df, level, fn)
```

That seems(!) to work. Finally we need a function with which we can compute the distances between all location available in the data set. Preferably it returns a distance matrix between all locations. I came up with the following:

```python
def location_dm(df, level='place', fn=jaccard_location, **kwargs):
    encode = encoder()
    locations = set(df[df[level] != '???'][level].values)
    dm = np.zeros((len(locations), len(locations)))
    for loc_a, loc_b in combinations(locations, 2):
        dist = fn(loc_a, loc_b, df, level=level, **kwargs)
        dm[encode(loc_a), encode(loc_b)] = dist
        dm[encode(loc_b), encode(loc_a)] = dist
    labels = sorted(locations, key=encode)
    return pd.DataFrame(dm, columns=labels, index=labels)
```

Both the Levenshtein distance and the Jaccard distance penalize mismatching motifs quite severe. However, some motifs are more related, more similar than other motifs. For example the motif D141.1 "Transformation: woman to bitch" is highly similar to D141 "Transformation: man to dog", yet it is penalized just as much as for the motif E601 "Reincarnation: former lives remembered" or any other motif.

To account for this I propose a new method to compute the distance between two motifs that incorporates the hierarchical relations between motifs in Thompson's *Motif-Index*. The distance between two motifs can be defined in terms of the shortest path length through the hierarchical tree between two motifs. The distance between D141 and D141.1 would then be 1.

Thompson's *Motif-Index* has been parsed and stored as an directed graph. I load the graph using the Python interface to [igraph](http://igraph.org/):

```python
import igraph

TMI = igraph.Graph.Read("../data/tmi.graphml", format="graphml")
TMI_ID = {node['id']: node.index for node in TMI.vs}
```

Using this graph we can define a function that returns the distance between motifs as follows:

```python
from phd.utils import memo

@memo
def weight(m1, m2, penalty=20):
    if m1[0] != m2[0]:
        return penalty
    try:
        return TMI.shortest_paths_dijkstra(
            TMI_ID[m1], TMI_ID[m2], weights='weight')[0][0]
    except KeyError:
        return penalty    
```

The update function in Levenshtein now becomes something like:

$$ \textrm{lev}_{a,b}(i, j) = \min \begin{cases} \textrm{lev}_{a,b}(i-1, j) + \delta \\ \textrm{lev}_{a, b}(i, j-1) + \delta \\ \textrm{lev}_{a, b}(i-1, j-1) + \begin{cases} SP(i-1, j-1) & if i-1_0 = j-1_0 \\ \delta & otherwise \end{cases}  \end{cases} $$

where $SP$ is the normalized Dijkstra shortest path length in Thompsons hierarchical *Motif-Index*. We will plug that into a new definition of levenshtein distance, as follows:

```python
from array import array

def weighted_leven(seq1, seq2, GAP=2, fn=weight):
    if seq1 == seq2: return 0.0
    if not seq1 or not seq2: return 1.0
    len1, len2 = len(seq1), len(seq2)
    if len1 < len2:
        len1, len2 = len2, len1
        seq1, seq2 = seq2, seq1
    column = array('f')
    for x in range(len2 + 1):
        column.append(x * GAP)
    length = array('f')
    for x in range(len2 + 1):
        length.append(x * GAP)
    
    for x in range(1, len1 + 1):
        column[0] = length[0] = x * GAP
        last = llast = (x-1) * GAP
        for y in range(1, len2 + 1):
            # distance
            old = column[y]
            insert = column[y - 1] + GAP
            delete = column[y] + GAP
            replace = last + fn(seq1[x - 1], seq2[y - 1])
            column[y] = min(insert, delete, replace)
            last = old
            # length
            lold = length[y]
            linsert = length[y - 1] + (GAP if insert == column[y] else 0)
            ldelete = length[y] + (GAP if delete == column[y] else 0)
            lreplace = llast + (fn(seq1[x - 1], seq2[y - 1]) 
                                if replace == column[y] else 0)
            length[y] = max(linsert, ldelete, lreplace)
            llast = lold
    return column[y] / float(length[y])
```

The following small test function shows that the function seem to work correctly...

```python
def test_leven():
    assert weighted_leven(
          ["B103", "D1601.5", "D1030"], ["B103", "D1601.5", "D1030"]) == 0
    assert weighted_leven(
           ["B103", "D1601.5", "D1030.1"], ["B103", "D1601.5", "D1030"]) == (1.0 / (3 * 2))
    assert weighted_leven(
           ["B103", "D1601.5", "D1030.1"], ["B103", "D1601", "D1030"]) == (2.0 / (3 * 2))
    assert weighted_leven(
           ["B103", "G303", "D1030.1"], ["B103", "D1601.5", "D1030"]) == \
           ((2 + 2 + 1) / (4. * 2))
    print 'all tests passed'
    
test_leven()
```

Before we will do that, we need to come up with a way to evaluate the resulting distance matrices (later we will also need a way to validate and evaluate the cluster solutions). One way to validate the distance matrices is by means of a metric called "Local incoherence". This metric computes the quality of a measurement of story differences. The idea behind the metric is that a location should be more similar to a location nearby than to a location a little further away. Local incoherence is defined as follows:

$$I_L = \frac{1}{n} \sum^n_{i=1} \frac{A_i - \hat{A_i}}{\hat{A_i}}$$

where

$$A_i = \sum^k_{j=1} d_{ij} \cdot 2^{jp}$$

and

$$\hat{A_i} = \sum^k_{j=1} \hat{d_{ij}} \cdot 2^{jp}$$

where $n$ is the number of locations, $d_{ij}$ is the geographical distance between location $i$ and location $j$ ordered by narrative distance, $\hat{d_{ij}}$ is the geographical distance between $i$ and $j$, ordered by geographical distance, $k$ is the upper limit for the number of dialectological neighborhood locations used and $p$ is a parameter that ???.

I will make use of the program `linc`, part of the dialectometrical software package [RuG/L04](http://www.let.rug.nl/~kleiweg/L04/) to compute this score:

```python
import sh, os

def linc(distance_file, coordinate_file):
    _linc = sh.Command(os.path.expanduser("~/local/RuG-L04/bin/linc"))
    return float(_linc("-L", distance_file, coordinate_file))

linc("dm.dif", "provinces.coo")
```

The `linc` program takes as input a distance file matrix in a particular [format](http://www.let.rug.nl/~kleiweg/L04/Manuals/formats.html#dif). The following function produces that format: 

```python
def to_diff_matrix_file(dm, filepath):
    "Write a distance matrix to the format required by RuG/L04."
    with open(filepath, 'w') as outfile:
        outfile.write("%s\n" % dm.shape[0])
        for label in dm.index:
            outfile.write('%s\n' % label)
        for i in range(1, dm.shape[0]):
            for j in range(i):
                outfile.write("%s\n" % 
                    (dm.values[i,j] if not np.isnan(dm.values[i,j]) else "NA"))
```

In order to get high quality topics, we want to filter words that are very uncommon and words that are very common. We'll define the following filter function:

```python
from collections import Counter
import numbers
from itertools import ifilterfalse

def filter_words(corpus, min_df=2, max_df=1.0):
    df = Counter(word for story in corpus for word in set(story))
        
    max_doc_count = (max_df if isinstance(max_df, numbers.Integral) 
                            else len(corpus) * max_df)
    min_doc_count = (min_df if isinstance(min_df, numbers.Integral)
                            else len(corpus) * min_df)

    filtered_words = set(word for word, count in df.iteritems() if
                         count > max_doc_count or count < min_doc_count)
    return lambda story: ifilterfalse(filtered_words.__contains__, story)
```

```python
def test_filter():
    string = 'abcdefghij'
    cps = [list(string[:i]) for i in range(1, len(string))]

    filter_ = filter_words(cps, min_df=1, max_df=1.0)
    assert set(c for d in cps for c in filter_(d)) == {'a','b','c','d','e','f','g','h','i'}
    
    filter_ = filter_words(cps, min_df=1, max_df=0.9)
    assert set(c for d in cps for c in filter_(d)) == {'b','c','d','e','f','g','h','i'}
    
    filter_ = filter_words(cps, min_df=2, max_df=0.8)
    assert set(c for d in cps for c in filter_(d)) == {'c','d','e','f','g','h'}

    print 'all tests passed'
    
test_filter()
```
